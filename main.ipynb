```python
# importing libraries
import pandas as pd
import numpy as np
```

```python
# loading the dataset
data = pd.read_csv('Data/Superstore.csv', encoding='ISO-8859-1')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
# any missing values?
data.isnull().sum()
```

```python
data.columns
```

```python
# analyze the performance of the retail business-
# dropping unnecessary columns like - row id, postal code
```

```python
data.drop(['Row ID', 'Postal Code'], axis = 1, inplace = True)
data.head()
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
```

```python
data.head()
```

```python
# shipping delay
data['Shipping Delay'] = (data['Ship Date'] - data['Order Date']).dt.days
```

```python
data.head(2)
```

```python
# profit margin
data['Profit Margin'] = data['Profit'] / data['Sales']
data['Profit Margin'] = data['Profit Margin'].replace([np.inf, -np.inf], np.nan)
```

```python
data.head(2)
```

```python
data.columns = data.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
data.rename(columns={'sub-category':'sub_category'}, inplace = True)
```

```python
data.to_csv('Output/cleaned_data.csv', index = False)
```

```python
from sqlalchemy import create_engine
```

```python
c_data = pd.read_csv('Output/cleaned_data.csv')
```

```python
#SQLite database engine
engine = create_engine('sqlite:///Output/retail.db', echo = False)
```

```python
c_data.to_sql('retail_sales', con = engine, if_exists = 'replace', index=False)
```

```python
from sqlalchemy import inspect
inspector = inspect(engine)
print(inspector.get_table_names())
# check
query = "SELECT * FROM retail_sales LIMIT 5"
preview = pd.read_sql(query, con = engine)
preview
```

```python
query = "SELECT * FROM retail_sales LIMIT 5"
preview = pd.read_sql(query, con = engine)
preview
```

```python
# Total sales and Profit by Region
query = "SELECT region, ROUND(SUM(sales), 2) AS total_sales, ROUND(SUM(profit), 2) AS total_profit FROM retail_sales GROUP BY region ORDER BY total_profit DESC"
preview = pd.read_sql(query, con = engine)
preview
```

```python
# Monthly Sales Trend
query = "SELECT strftime('%Y-%m', order_date) AS order_month, ROUND(SUM(sales), 2) AS monthly_sales FROM retail_sales GROUP BY order_month ORDER BY order_month"
preview = pd.read_sql(query, con = engine)
preview
```

```python
# Top 10 Profitable customers
query = "SELECT customer_name, ROUND(SUM(profit), 2) AS total_profit FROM retail_sales GROUP BY customer_name ORDER BY total_profit DESC LIMIT 10"
preview = pd.read_sql(query, con = engine)
preview
```

```python
# Average shipping delay by region
query = "SELECT region, ROUND(AVG(shipping_delay), 2) AS avg_shipping_delay FROM retail_sales GROUP BY region ORDER BY avg_shipping_delay DESC"
preview = pd.read_sql(query, con = engine)
preview
```

```python
# Profit margin by category
query = "SELECT category, ROUND(SUM(profit) / SUM(sales), 2) AS avg_profit_margin FROM retail_sales GROUP BY category ORDER BY avg_profit_margin DESC"
preview = pd.read_sql(query, con = engine)
preview
```

```python
# Sales & Profit by category and sub-category
query = "SELECT category, sub_category, ROUND(SUM(sales), 2) AS total_sales, ROUND(SUM(profit), 2) AS total_profit FROM retail_sales GROUP BY category, sub_category ORDER BY total_profit DESC"
preview = pd.read_sql(query, con = engine)
preview
```

```python
# Customer Segment Performance
query = '''SELECT segment,
       ROUND(SUM(sales), 2) AS total_sales,
       ROUND(SUM(profit), 2) AS total_profit
FROM retail_sales
GROUP BY segment'''
preview = pd.read_sql(query, con = engine)
preview
```

```python
# Top discounted products
query = '''SELECT product_name, discount, profit
FROM retail_sales
WHERE discount > 0 AND profit > 0
ORDER BY discount DESC
LIMIT 10'''
preview = pd.read_sql(query, con = engine)
preview
```

```python
# Orders with negative profit
query = '''SELECT order_id, product_name, region, profit
FROM retail_sales
WHERE profit < 0
ORDER BY profit ASC
LIMIT 10'''
preview = pd.read_sql(query, con = engine)
preview
```

```python
# Repeat customers - more than 3 orders
query = '''SELECT customer_name, COUNT(order_id) AS total_orders
FROM retail_sales
GROUP BY customer_name
HAVING total_orders > 3
ORDER BY total_orders DESC'''
preview = pd.read_sql(query, con = engine)
preview
```
